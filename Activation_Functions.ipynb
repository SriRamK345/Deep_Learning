## Sigmoid

```python
import numpy as np

def sigmoid(x):
    s = 1 / (1 + np.exp(-x)) # sigmoid function
    ds = s*(1-s) # derivative
    return s, ds
```

```python
print(sigmoid(0))
print(sigmoid(np.array([0,1,2])))
print(sigmoid(-1))  # outpout always in postive
```

```python
import matplotlib.pyplot as plt

X = np.arange(-10,10,0.1)
# Y = sigmoid(X)[0] #
# plt.plot(X,Y)
# plt.show()

fig, ax = plt.subplots(figsize = (10,6))
ax.plot(X,sigmoid(X)[0])
ax.plot(X,sigmoid(X)[1])
plt.show()
```

## Tanh

```python
def tanh(x):
  t = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
  # t = np.tanh(x)
  dt = 1 - t**2
  return t, dt
```

```python
X = np.arange(-10,10,0.1)

fig, ax = plt.subplots(figsize = (10,6))
ax.plot(X,tanh(X)[0])
ax.plot(X,tanh(X)[1])
plt.show()
```

## ReLU

```python
def relu(x):
  r = np.maximum(0,x)
  dr = np.where(x>0,1,0)
  return r, dr
```

```python
X= np.arange(-10,10,0.1)

fig, ax = plt.subplots(figsize = (10,6))
ax.plot(X,relu(X)[0])
ax.plot(X,relu(X)[1])
plt.show()
```
